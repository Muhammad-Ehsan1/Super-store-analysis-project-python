# DATA VISUALIZATION

```python
sns.lineplot(x=df["year"],y=df["Sales"],label= "sales",marker="o")
plt.title("Sales by year")
plt.grid()
plt.show()
```

```python
sns.lineplot(x=df["year"],y=df["Profit"],label= "Profit",marker="*",color = "green")
plt.title("Profit by year")
plt.grid()
plt.show()
```

```python
sns.lineplot(x=df["month"],y=df["Sales"],label= "sales",marker="o")
sns.lineplot(x=df["month"],y=df["Profit"],label= "Profit",marker="*",color = "green")
plt.title("Sales & Profit by Months")
plt.grid()
plt.show()
```

```python
ax=sns.barplot(data= df , x = "Category",y = "Sales",hue="Category",palette="tab20")
for containers in ax.containers:
    ax.bar_label(containers)
plt.title("Sales by Category")
plt.show()
```

```python
by=sns.barplot(data= df , x = "Category",y = "Profit",hue="Category")
for contain in by.containers:
    by.bar_label(contain)
plt.title("Profit by Category")
plt.show()
```

```python
plt.figure(figsize=(7,7))
my=sns.barplot(data= df , x = "Sales" ,y = "Sub-Category",hue="Sub-Category",palette="tab20")
for containers in my.containers:
    my.bar_label(containers)
plt.title("Sales by sub-Category")
plt.show()
```

```python
mq=sns.barplot(data= df , x = "Profit" ,y = "Sub-Category",hue="Sub-Category")
for containers in mq.containers:
    mq.bar_label(containers)
plt.title("Profit by Sub-Category")
plt.show()
```

```python
rs=sns.countplot(x="Segment",data=df,hue="Segment")
for containers in rs.containers:
    rs.bar_label(containers)
plt.title("Orders per segment")
plt.show()
```

```python
df["Dilivary_time"]=df["Ship Date"]-df["Order Date"]
sns.boxplot(x= "Dilivary_time",data=df)
plt.title("Delivery Time")
plt.show()
```

```python
rb=sns.barplot(x=df["Ship Mode"], y=df["Sales"].mean(),hue=df["Ship Mode"])
for contain in rb.containers:
    rb.bar_label(contain)
plt.title("Shipping Mode vs Average Sales")
plt.show()
```

```python
sns.regplot(data =df, y ="Discount",x="Profit")
plt.title("Discount vs Profit Relationship")
plt.show()
```

```python
sns.violinplot(data = df , x= "Segment",y = "Sales")
plt.title("Customer Segment Purchase Behavior")
plt.show()
```

```python
corr_matrix = df[['Sales', 'Profit', 'Discount', 'Quantity']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Sales, Profit, Discount, and Quantity')
plt.show()
```
